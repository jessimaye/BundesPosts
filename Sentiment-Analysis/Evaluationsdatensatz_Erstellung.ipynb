```python
import pandas as pd
import random
```

```python
df = pd.read_csv("global_posts.csv")
```

```python
df.index = df.index + 1
df
```

```python
#for annotation unnecessary columns are dropped
df_mini = df.drop(['Unnamed: 0', 'thread_id', 'parent_id', 'author', 'author_avatar_url', 'timestamp', 'type', 'url', 'image_url', 'media_url', 'num_likes', 'num_comments', 'num_media', 'location_name', 'location_latlong', 'location_city', 'unix_timestamp', 'media_n', 'media_filename'], axis=1)
df_mini
```

```python
#2000 random entries are choosen for the evaluation dataframe
e_dict = {}

while len(e_dict)<2000:
    random_number = random.randrange(1, len(df_mini))
    e_dict[random_number] = df_mini.iloc[random_number-1]

e_df = pd.DataFrame(e_dict)
e_df = e_df.transpose()

e_df
```
